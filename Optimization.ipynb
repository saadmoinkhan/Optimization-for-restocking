# Optimization for restocking

#### Optimization using Excel for a company which needs multiple items for restocking. They have different suppliers bidding various items at their price point. 

#### This model will run all combinations to find out the optimized solution which is filling the requirement with lowest cost.

The data is abitrary but the technique can be used by any organization/company.

We will create new columns:-
* <b>Obtained</b> - This column contains the obtained units of products after ordering from different suppliers.</br>
Formula used = SUMPRODUCT(C4:N4,$C$20:$N$20) and then dragged on column</br></br>
* <b>Extra</b> - This column contains the extra items that remain after filling the requirement.</br>
Formula used = Q4-O4 and then dragged on column</br></br>
* <b>Decisions</b> - This row contains binary decision, 1/0, which signify whether the bid was taken or not.</br></br>
* <b>Cost</b> - Cost is calulated by adding the different bids taken.</br>
Formula used = SUMPRODUCT(C18:N18,C20:N20)

We will now use solver to minimize the Cost by changing the decision cells, and then adding the constraints like Obtained must be equal or greater than Required.

And solver optimizes the cost, although we are getting some extra units of different items, but this is the best solution to minimize the Cost.
